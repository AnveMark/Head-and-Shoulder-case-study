```python
import pandas as pd
from scipy.stats import chi2_contingency

# Build contingency table
contingency_table = pd.crosstab(
    df_skiplogic['Q1 How would you describe your dandruff condition'],
    df_skiplogic['Q2 Do you experience frequent hair fall?']
)

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square test result:")
print(f"Chi2 statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")
```

```python
from scipy.stats import kruskal

# Step 1: Map ordinal values in Q7 to numeric scale (adjust mapping as per actual labels)
hair_fall_mapping = {
    'Low': 1,
    'Moderate': 2,
    'High': 3,
    'Always': 4
}
df_skiplogic['Q7_numeric'] = df_skiplogic['Q7 What is the frequency of your hair fall?'].map(hair_fall_mapping)

# Step 2: Group Q7_numeric by dandruff severity (Q1)
grouped = df_skiplogic.dropna(subset=['Q1 How would you describe your dandruff condition', 'Q7_numeric'])\
    .groupby('Q1 How would you describe your dandruff condition')['Q7_numeric'].apply(list)

# Step 3: Perform Kruskal-Wallis test
stat, p = kruskal(*grouped)

print("Kruskal–Wallis test result:")
print(f"H statistic: {stat:.4f}")
print(f"P-value: {p:.4f}")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Step 1: Standardize treatment column (Q32)
df_final['Q32'] = df_final['Q32 Which product do you prefer: Test or Control?'].str.strip().str.capitalize()

# Step 2: Map treatment assignment
df_final['Treatment'] = df_final['Q32'].map({'Test': 1, 'Control': 0})

# Step 3: Define required columns for uplift modeling
required_cols = [
    'Treatment',
    'Q36. How likely are you to purchase the preferred shampoo?',
    'Q24 How would you rate the quality of the packaging?',
    'Q27 If this shampoo changed only its packaging, how likely are you to try it again?',
    'Q30 How would you rate this brand overall on a scale of 1 to 10?',
    'Q35_VfM'
]

# Step 4: Drop rows with any missing values
df_model = df_final[required_cols].dropna()

# Step 5: Rename target for simplicity
df_model = df_model.rename(columns={
    'Q36. How likely are you to purchase the preferred shampoo?': 'Q36',
    'Q24 How would you rate the quality of the packaging?': 'Q24',
    'Q27 If this shampoo changed only its packaging, how likely are you to try it again?': 'Q27',
    'Q30 How would you rate this brand overall on a scale of 1 to 10?': 'Q30'
})

# Step 6: Convert target to numeric
df_model['Q36'] = pd.to_numeric(df_model['Q36'], errors='coerce')
df_model = df_model.dropna()

# Step 7: Split into treatment and control
features = ['Q24', 'Q27', 'Q30', 'Q35_VfM']
X_treat = df_model[df_model['Treatment'] == 1][features]
y_treat = df_model[df_model['Treatment'] == 1]['Q36']

X_control = df_model[df_model['Treatment'] == 0][features]
y_control = df_model[df_model['Treatment'] == 0]['Q36']

# Step 8: Sanity check
print(f"✅ Treatment samples: {len(X_treat)}, Control samples: {len(X_control)}")

# Step 9: Fit two separate models if both groups have data
if len(X_treat) > 0 and len(X_control) > 0:
    model_treat = RandomForestRegressor(random_state=42)
    model_control = RandomForestRegressor(random_state=42)

    model_treat.fit(X_treat, y_treat)
    model_control.fit(X_control, y_control)

    # Predict for all
    X_all = df_model[features]
    pred_treat = model_treat.predict(X_all)
    pred_control = model_control.predict(X_all)

    # Compute uplift: E[Y1 - Y0]
    df_model['Uplift'] = pred_treat - pred_control
    average_uplift = df_model['Uplift'].mean()

    print(f"\n✅ Estimated average uplift in Purchase Intent (Q36): {average_uplift:.3f}")
else:
    print("❌ Not enough data in one or both groups for uplift modeling.")
```

```python
# Load spaCy model
nlp = spacy.load('en_core_web_sm', disable=['ner', 'parser'])

# Cleaning + lemmatization function
def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", "", text)  # HTML tags
    text = re.sub(r"[^a-z\s]", "", text)  # punctuation and numbers
    text = re.sub(r"\s+", " ", text).strip()  # extra spaces
    return text

def lemmatize(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return " ".join(tokens)

# Apply cleaning and lemmatization
text_df['Cleaned_Text'] = text_df['Raw_Text'].progress_apply(clean_text).progress_apply(lemmatize)
```

```python
# Load Sentence-BERT model
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode cleaned text
text_df['Embedding'] = text_df['Cleaned_Text'].progress_apply(lambda x: sbert_model.encode(x))

# Convert embeddings to numpy matrix
embedding_matrix = np.vstack(text_df['Embedding'].values)

# Save
np.save('embedding_matrix.npy', embedding_matrix)
text_df[['id', 'Question_ID']].to_csv('embedding_mapping.csv', index=False)
```

```python
import pandas as pd
from scipy.stats import f_oneway, kruskal

# Assumes merged_df has columns: ['id', 'Aspect', 'Sentiment_Score', 'Age', 'Gender', 'Q6 / Q7 NCCS']
group_var = 'Q6 / Q7 NCCS'  # Change to 'Age' or 'Gender' as needed

aspects = merged_df['Aspect'].unique()
test_results = []

for aspect in aspects:
    subset = merged_df[merged_df['Aspect'] == aspect]
    grouped = subset.groupby(group_var)['Sentiment_Score'].apply(list)

    if len(grouped) > 1:
        try:
            # Prepare groups for tests
            score_lists = [scores for scores in grouped if len(scores) > 1]  # drop groups with only 1 value

            if len(score_lists) >= 2:  # need at least 2 groups with data
                anova_stat, anova_p = f_oneway(*score_lists)
                kruskal_stat, kruskal_p = kruskal(*score_lists)

                test_results.append({
                    'Aspect': aspect,
                    'Group_By': group_var,
                    'ANOVA_p': anova_p,
                    'Kruskal_p': kruskal_p
                })
        except Exception as e:
            print(f"[Skipped] {aspect} due to: {e}")

# Convert to DataFrame
test_df = pd.DataFrame(test_results)

# Optional: filter significant results
if not test_df.empty:
    significant = test_df[(test_df['Kruskal_p'] < 0.05) | (test_df['ANOVA_p'] < 0.05)]
    print("Significant differences found:")
    print(significant)
else:
    print("No valid test results were computed. Please check data.")

# Display full results
print("\nAll test results:")
print(test_df)

# Optional: Save
test_df.to_csv("aspect_sentiment_significance.csv", index=False)
```

```python
def get_top_keywords_by_segment(df, segment_col, text_col, top_n=10):
    result = {}
    for segment in df[segment_col].dropna().unique():
        subset = df[df[segment_col] == segment]
        tfidf = TfidfVectorizer(max_features=1000)
        tfidf_matrix = tfidf.fit_transform(subset[text_col])
        tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
        top_keywords = tfidf_df.sum().sort_values(ascending=False).head(top_n)
        result[segment] = top_keywords
    return result
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, precision_recall_curve

def plot_curves(y_true, probs, model_name):
    fpr, tpr, _ = roc_curve(y_true, probs)
    precision, recall, _ = precision_recall_curve(y_true, probs)

    plt.figure(figsize=(12, 5))

    # ROC Curve
    plt.subplot(1, 2, 1)
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {roc_auc_score(y_true, probs):.2f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f"{model_name} - ROC Curve")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()

    # Precision-Recall Curve
    plt.subplot(1, 2, 2)
    plt.plot(recall, precision, label=model_name)
    plt.title(f"{model_name} - Precision-Recall Curve")
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_curves(y_test, rf_probs, "Random Forest")
plot_curves(y_test, xgb_probs, "XGBoost")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score
import shap

# -------------------------------
# Step 1: Load and Clean Data
# -------------------------------

# Assuming df_final is already loaded

# Replace invalid/placeholder strings with np.nan
df_final.replace(["N/A", "Not Sure", "", "na", "NA", "missing"], np.nan, inplace=True)

# Define predictors and target from your column list
predictor_cols = [
    'Q4 Respondent Gender',
    'Q5 Please select the age range you fall into.',
    'location' if 'location' in df_final.columns else None,
    'Q1 How would you describe your dandruff condition',
    'Q2 Do you experience frequent hair fall?',
    'Q3 Do you experience scalp itchiness?',
    'Q4 Do you have an oily scalp?',
    'Q5 How long does your hair stay fresh before becoming oily again?',
    'Q6 Do you face any hair fall issues currently?',
    'Q7 What is the frequency of your hair fall?',
    'Q8 When do you usually notice your hair fall?',
    'Q9 Is your scalp sensitive?',
    'Q10 Is your scalp dry?',
    'Q17 How strongly do discounts or promotions influence your shampoo purchase?',
    'Q24 How would you rate the quality of the packaging?',
    'Q30 How would you rate this brand overall on a scale of 1 to 10?',
    'Q34. How premium did the preferred product look?',
    'Q35. Was it value for money',
    'Q36. How likely are you to purchase the preferred shampoo?',
    'Q38. How similar were the test and control products?',
    'Brand Recall Score (out of 10)',
    'Fixation on Product Claims (%)',
    'Price Visibility Recognition (seconds)',
    'Dwell Time on Brand Logo (ms)',
    'Dwell Time on Product Image (ms)',
    'Dwell Time on Text (ms)',
    'Peripheral Vision Engagement (%)',
    'Ad Appeal Rating (1-10)',
    'Association with Luxury (1-5)',
    'Visual Memorability Score (1-10)',
    'Likelihood to Click Ad (1-5)',
    'scalp_severity_score',
    'hair_fall_frequency_index',
    'packaging_engagement_flag'
]

# Filter only existing columns
predictor_cols = [col for col in predictor_cols if col in df_final.columns]
target_col = 'purchase_intent_binary'

# Drop rows where target is missing
df_final = df_final.dropna(subset=[target_col])

# Create feature and target sets
X = df_final[predictor_cols]
y = df_final[target_col]

# -------------------------------
# Step 2: Train/Test Split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# -------------------------------
# Step 3: Preprocessing Pipeline
# -------------------------------
numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# -------------------------------
# Step 4: Modeling Pipeline
# -------------------------------
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42, n_estimators=100))
])

# -------------------------------
# Step 5: Fit Model
# -------------------------------
model_pipeline.fit(X_train, y_train)

# -------------------------------
# Step 6: Evaluate Model
# -------------------------------
y_proba = model_pipeline.predict_proba(X_test)[:, 1]
threshold = 0.3  # tuned threshold
y_pred = (y_proba >= threshold).astype(int)

print(f"ROC AUC: {roc_auc_score(y_test, y_proba):.3f}")
print(f"Precision: {precision_score(y_test, y_pred):.3f}")
print(f"Recall: {recall_score(y_test, y_pred):.3f}")
print(f"F1 Score: {f1_score(y_test, y_pred):.3f}")

# -------------------------------
# Step 7: SHAP Explainability
# -------------------------------
rf_model = model_pipeline.named_steps['classifier']
X_train_transformed = model_pipeline.named_steps['preprocessor'].transform(X_train)

# Convert to dense if sparse
if hasattr(X_train_transformed, "todense"):
    X_train_transformed = X_train_transformed.todense()

explainer = shap.Explainer(rf_model, X_train_transformed)
shap_values = explainer(X_train_transformed[:100])  # Limit for speed

# Plot summary
shap.summary_plot(shap_values, X_train_transformed[:100])
```
